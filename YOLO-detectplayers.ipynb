```python
from ultralytics import YOLO

# Create a new YOLO model from scratch
model = YOLO('yolov8n.yaml')

# Load a pretrained YOLO model (recommended for training)
model = YOLO('yolov8n.pt')
```

```python
import os
os.listdir(frames_dir)
```

```python
def yolo_detect_players(frames_dir,dect_num):
    for filename in os.listdir(frames_dir):
        img_path = os.path.join(frames_dir, filename)
        
        #specify detect class and result number
        model.predict(img_path, save=True, classes=[0], max_det = dect_num)
```

```python
single_frames_dir = './SINGLEframes-fullcourt/SINGLEframes-fullcourt'
yolo_detect_players(single_frames_dir,2)
```

```python
double_frames_dir = './DOUBLEframes-fullcourt/DOUBLEframes-fullcourt'
yolo_detect_players(double_frames_dir,4)
```
